```python
from sklearn.cluster import KMeans
from dataloader import Trial
import cv2
import numpy as np
```

```python
video_path = r'..\data\TH8\TH8ST17A3D250TS0_2.mov'
recording = Trial(video_path)
recording.crop_top(inplace=True)
```

```python


img = recording[0].copy()

Z = img.reshape((-1, 3))
Z = np.float32(Z)

criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 4
_, labels, centers = cv2.kmeans(Z, K, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

centers = np.uint8(centers)
quantized = centers[labels.flatten()]
quantized = quantized.reshape(img.shape)

quantized = cv2.medianBlur(quantized, 5)

cv2.imshow("Original", img)
cv2.imshow("Quantized", quantized)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
